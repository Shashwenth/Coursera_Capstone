```python
maps2={}
maps2[city] = folium.Map(location=[latitude,longitude], zoom_start=11)
venues_mean_coor = [df_venues[city]['Lat'].mean(), df_venues[city]['Lng'].mean()] 
    # add markers to map
for lat, lng, label in zip(df_venues[city]['Lat'], df_venues[city]['Lng'], df_venues[city]['Name']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
            [lat, lng],
            radius=5,
            popup=label,
            color='blue',
            fill=True,
            fill_color='#3186cc',
            fill_opacity=0.7,
            parse_html=False).add_to(maps2[city])
    folium.PolyLine([venues_mean_coor, [lat, lng]], color="green", weight=1.5, opacity=0.5).add_to(maps2[city])
    
label = folium.Popup("Mean Co-ordinate", parse_html=True)
folium.CircleMarker(
        venues_mean_coor,
        radius=10,
        popup=label,
        color='green',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(maps2[city])

print(city)
print("Mean Distance from Mean coordinates")
print(np.mean(np.apply_along_axis(lambda x: np.linalg.norm(x - venues_mean_coor),1,df_venues[city][['Lat','Lng']].values)))
```

```python
q={}
R=6373.0
r=[]
for i in venue_loc1:
    
    w={}
    e=venue_loc1[i]
    la1=math.radians(e['venue.location.lat'])
    lo1=math.radians(e['venue.location.lng'])
    
    for j in venue_loc:
        
        s=venue_loc[j]
        la2=math.radians(s['venue.location.lat'])
        lo2=math.radians(s['venue.location.lng'])
        dla=la1-la2
        dlo=lo1-lo2
        z = math.sin(dla / 2)**2 + math.cos(la1) * math.cos(la2) * math.sin(dlo / 2)**2
        x = 2 * math.atan2(math.sqrt(z), math.sqrt(1 - z))
        dist=R*x
        w[j]=round(dist,2)
    r.append([i,w])
    
        
        
        
```
